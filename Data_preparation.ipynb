```python
df = df.drop('Order', axis=1)
df = df.drop('Crude rate', axis=1)
# Remove the first row
df = df.drop(df.index[0])
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df
# Transpose the DataFrame for better visualization
#df_transposed = df.T

# Set up the plot
plt.figure(figsize=(15, 10))

# Create a heatmap with y-axis as cancer sites and x-axis as age
sns.heatmap(df, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)

# Set labels and title
plt.ylabel('Cancer Sites')
plt.xlabel('Age')
plt.title('Crude Rates Heatmap')

# Show the plot
plt.show()
```

```python
# make the small values more sensitive

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Assuming you have a DataFrame named df

# Set up the plot
plt.figure(figsize=(15, 10))

# Create a heatmap with y-axis as cancer sites and x-axis as age
sns.heatmap(df, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5, norm=LogNorm())

# Set labels and title
plt.ylabel('Cancer Sites')
plt.xlabel('Age')
plt.title('Crude Rates Heatmap (Logarithmic Scale)')

# Show the plot
plt.show()
```

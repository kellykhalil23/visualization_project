```python
import pandas as pd

# Function to rank the top 5 cancer sites for each age group
def rank_top_cancers(data):
    # Create an empty list to store the rankings
    rankings_list = []

    # Loop through each age group
    for age_group in data.columns[1:]:
        # Sort the cancer sites based on crude rate values in descending order
        sorted_cancers = data[['Cancer site', age_group]].sort_values(by=age_group, ascending=False)

        # Take the top 5 cancer sites
        top_5_cancers = sorted_cancers.head(5)['Cancer site'].tolist()

        # Append the rankings to the list
        rankings_list.append({'AgeGroup': age_group, '1st': top_5_cancers[0], '2nd': top_5_cancers[1],
                              '3rd': top_5_cancers[2], '4th': top_5_cancers[3], '5th': top_5_cancers[4]})

    # Create a DataFrame from the list
    rankings = pd.DataFrame(rankings_list)

    return rankings

# Assuming your data is stored in a DataFrame named 'dfSUIS'
rankings = rank_top_cancers(dfSUIS)

# Display the rankings
rankings


```

```python
import pandas as pd

def rank_top_cancers_combined(dataframes):
    # Create an empty list to store the rankings
    rankings_list = []

    # Loop through each dataframe in the list
    for df in dataframes:
        # Loop through each age group in the dataframe
        for age_group in df.columns[1:]:
            # Sort the cancer sites based on crude rate values in descending order
            sorted_cancers = df[['Cancer site', age_group]].sort_values(by=age_group, ascending=False)

            # Take the top 5 cancer sites
            top_5_cancers = sorted_cancers.head(5)['Cancer site'].tolist()

            # Append the rankings to the list
            rankings_list.append({'AgeGroup': age_group, '1st': top_5_cancers[0], '2nd': top_5_cancers[1],
                                  '3rd': top_5_cancers[2], '4th': top_5_cancers[3], '5th': top_5_cancers[4],
                                  'Dataset': df.name})  # Add the dataset name

    # Create a DataFrame from the list
    rankings = pd.DataFrame(rankings_list)

    return rankings

# Assuming your dataframes are named dfSUIS, dfUSA, dfCANADA, and dfgermany
# Add names to the dataframes
dfSUIS.name = 'SUIS'
dfUSA.name = 'USA'
dfCANADA.name = 'CANADA'
dfgermany.name = 'Germany'

# Combine the rankings from multiple dataframes
combined_rankings = rank_top_cancers_combined([dfSUIS, dfUSA, dfCANADA, dfgermany])

# Display the combined rankings
combined_rankings
```
